## Use Seaborn library to draw a Gaussian Normal Distribution of “AGE” feature in Boston Housing Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from sklearn.datasets import load_boston
dataset=load_boston()
dataframe=pd.DataFrame(data=dataset['data'],columns=dataset['feature_names'])
dataframe['target y'] = dataset['target']
dataframe.head()
```

```python
sns.distplot(dataframe['AGE'])
```

```python
from scipy.stats import norm
x=norm.fit(dataframe['AGE'])
sns.distplot(x)
```

```python
sns.distplot(dataframe['AGE'],hist=False)
```

```python
sns.distplot(dataframe['AGE'],hist=False,fit=norm)
```
